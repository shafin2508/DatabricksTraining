How is the dataset indexed?

Which was the most-ordered item?

```
from pyspark.sql.functions import col

# Cast the "quantity" column to integer
df = df.withColumn("quantity", col("quantity").cast('int'))

# Group by "item_name" and sum "quantity"
a = df.groupBy("item_name").sum("quantity").orderBy("sum(quantity)", ascending=False)

a.select("item_name").show(1)
```

For the most-ordered item, how many items were ordered?

```
a.show(1)
```

What was the most ordered item in the choice_description column?

```
import pyspark.sql.functions 

df = df.withColumn("quantity", col("quantity").cast('int'))

a = df.groupBy("choice_description").sum("quantity").orderBy("sum(quantity)", ascending=False).dropna()
a = a.filter(col("choice_description") != "NULL")

a.select("choice_description").show(1)
```

Create a lambda function and change the type of item price

Check the item price type

```
from pyspark.sql import Row

# Convert DataFrame to RDD
rdd = df.rdd

# Use map to change the type of item_price to float, removing the dollar sign before conversion
rdd_transformed = rdd.map(lambda row: Row(**{k: (float(v.replace('$', '')) if k == 'item_price' else v) for k, v in row.asDict().items()}))

# Convert RDD back to DataFrame
df_transformed = spark.createDataFrame(rdd_transformed)

# Show the schema to check the type of item_price
df_transformed.printSchema()
```

How much was the revenue for the period in the dataset?

```
import pyspark.sql.functions as F

df_transformed = df_transformed.withColumn("item_price", F.col("item_price").cast("float"))

df_transformed.select(F.sum("item_price")).show()
```

How many orders were made in the period?

```
df.select("order_id").distinct().count()
```

What is the average revenue amount per order?

```
revenue = df_transformed.withColumn("item_price", F.col("item_price").cast("float"))
revenue=revenue.withColumn("TotalPrice", col("quantity") * col("item_price"))
revenue = revenue.groupBy("order_id").agg(F.avg("TotalPrice").alias("avg_revenue"))

display(revenue)
```
